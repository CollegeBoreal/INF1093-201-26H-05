## 1. Installation de Miniforge via Chocolatey

Pour installer Miniforge sous Windows PowerShell (en tant qu'administrateur) :

```powershell
choco install miniforge3 -y
```

Le chemin d'installation par défaut est : `C:\tools\miniforge3`

## 2. Configuration du PATH PowerShell

Ajout de Miniforge au profil PowerShell (`$PROFILE`) :

```powershell
# >>> Miniforge3 initialization >>>
$env:Path += ";C:\tools\miniforge3;C:\tools\miniforge3\Scripts;C:\tools\miniforge3\Library\bin"
# <<< Miniforge3 initialization <<<

# >>> conda initialize >>>
(& "C:\tools\miniforge3\Scripts\conda.exe" "shell.powershell" "hook") | Out-String | Invoke-Expression
# <<< conda initialize <<<
```

Recharger le profil :
```powershell
. $PROFILE
```

## 3. Création de l'environnement conda

```powershell
conda create -n INF1093 python=3.12 -y
conda activate INF1093
```

## 4. Installation de JupyterLab

```powershell
conda install -c conda-forge jupyterlab -y
jupyter lab
```
