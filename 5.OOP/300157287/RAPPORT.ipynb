```python


# =========================
# IMPORTS
# =========================
import math
import numpy as np
import matplotlib.pyplot as plt
import os

# Créer dossier images
if not os.path.exists("images"):
    os.makedirs("images")

# =========================
# CLASSES
# =========================
class Figure:
    def __init__(self, nom):
        self.nom = nom

    def aire(self):
        return 0

    def afficher_info(self):
        return f"{self.nom}"

class Carre(Figure):
    def __init__(self, cote):
        super().__init__("Carré")
        self.cote = cote

    def aire(self):
        return self.cote ** 2

class Cercle(Figure):
    def __init__(self, rayon):
        super().__init__("Cercle")
        self.rayon = rayon

    def aire(self):
        return math.pi * self.rayon ** 2

class Triangle(Figure):
    def __init__(self, base, hauteur):
        super().__init__("Triangle")
        self.base = base
        self.hauteur = hauteur

    def aire(self):
        return (self.base * self.hauteur) / 2

# =========================
# STYLE GLOBAL
# =========================
plt.rcParams.update({
    "figure.figsize": (6,6),
    "axes.titlesize": 14,
    "axes.labelsize": 12
})

# =========================
# AFFICHAGE CARRE
# =========================
def afficher_carre(carre):
    c = carre.cote

    x = [0, c, c, 0, 0]
    y = [0, 0, c, c, 0]

    plt.figure()
    plt.plot(x, y, linewidth=2)
    plt.fill(x, y, alpha=0.2)

    # mesures
    plt.text(c/2, -0.4, f"{c}", ha='center', fontsize=11)
    plt.text(-0.4, c/2, f"{c}", va='center', rotation=90, fontsize=11)

    # aire
    plt.text(c/2, c/2, f"Aire\n{carre.aire()}", ha='center', fontsize=12)

    plt.title("Carré")
    plt.axis('equal')
    plt.xlim(-1, c+1)
    plt.ylim(-1, c+1)
    plt.grid(alpha=0.3)

    plt.savefig("images/carre.png", dpi=300)
    plt.show()

# =========================
# AFFICHAGE CERCLE
# =========================
def afficher_cercle(cercle):
    r = cercle.rayon
    theta = np.linspace(0, 2*np.pi, 300)

    x = r * np.cos(theta)
    y = r * np.sin(theta)

    plt.figure()
    plt.plot(x, y, linewidth=2)
    plt.fill(x, y, alpha=0.2)

    # rayon
    plt.plot([0, r], [0, 0], linestyle='--')
    plt.text(r/2, 0.3, f"r = {r}", ha='center')

    # aire
    plt.text(0, 0, f"Aire\n{cercle.aire():.2f}", ha='center')

    plt.title("Cercle")
    plt.axis('equal')
    plt.xlim(-r-1, r+1)
    plt.ylim(-r-1, r+1)
    plt.grid(alpha=0.3)

    plt.savefig("images/cercle.png", dpi=300)
    plt.show()

# =========================
# AFFICHAGE TRIANGLE
# =========================
def afficher_triangle(triangle):
    b = triangle.base
    h = triangle.hauteur

    x = [0, b, 0, 0]
    y = [0, 0, h, 0]

    plt.figure()
    plt.plot(x, y, linewidth=2)
    plt.fill(x, y, alpha=0.2)

    # base
    plt.text(b/2, -0.4, f"b = {b}", ha='center')

    # hauteur
    plt.plot([0, 0], [0, h], linestyle='--')
    plt.text(-0.4, h/2, f"h = {h}", rotation=90, va='center')

    # aire
    plt.text(b/3, h/3, f"Aire\n{triangle.aire()}", ha='center')

    plt.title("Triangle")
    plt.axis('equal')
    plt.xlim(-1, b+1)
    plt.ylim(-1, h+1)
    plt.grid(alpha=0.3)

    plt.savefig("images/triangle.png", dpi=300)
    plt.show()

# =========================
# MAIN
# =========================
def main():
    c1 = Carre(4)
    c2 = Cercle(3)
    t1 = Triangle(5, 3)

    print(c1.afficher_info(), "Aire =", c1.aire())
    print(c2.afficher_info(), "Aire =", c2.aire())
    print(t1.afficher_info(), "Aire =", t1.aire())

    afficher_carre(c1)
    afficher_cercle(c2)
    afficher_triangle(t1)

    print("\nPolymorphisme :")
    for f in [c1, c2, t1]:
        print(f"{f.nom} → Aire: {f.aire()}")

if __name__ == "__main__":
    main()
```
