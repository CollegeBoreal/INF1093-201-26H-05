```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
```

## Visualisation des 12 formes 2D

Cette section affiche toutes les formes 2D du rapport.


```python

import math

def config_axes_2d(ax, titre):
    ax.set_title(titre)
    ax.set_aspect('equal')
    ax.grid(True)


def polygone_regulier_points(n, rayon, rotation=0.0):
    angles = np.linspace(0, 2*np.pi, n, endpoint=False) + rotation
    x = rayon * np.cos(angles)
    y = rayon * np.sin(angles)
    return np.append(x, x[0]), np.append(y, y[0])


def afficher_carre(carre):
    cote = carre.cote
    x = [0, cote, cote, 0, 0]
    y = [0, 0, cote, cote, 0]
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.plot(x, y)
    ax.fill(x, y, alpha=0.3)
    config_axes_2d(ax, f"Carré — côté={cote}, aire={carre.aire():.2f}")
    plt.show()


def afficher_rectangle(rectangle):
    largeur = rectangle.largeur
    hauteur = rectangle.hauteur
    x = [0, largeur, largeur, 0, 0]
    y = [0, 0, hauteur, hauteur, 0]
    fig, ax = plt.subplots(figsize=(6, 4))
    ax.plot(x, y)
    ax.fill(x, y, alpha=0.3)
    config_axes_2d(ax, f"Rectangle — largeur={largeur}, hauteur={hauteur}, aire={rectangle.aire():.2f}")
    plt.show()


def afficher_triangle(triangle):
    base = triangle.base
    hauteur = triangle.hauteur
    x = [0, base, base / 2, 0]
    y = [0, 0, hauteur, 0]
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.plot(x, y)
    ax.fill(x, y, alpha=0.3)
    config_axes_2d(ax, f"Triangle — base={base}, hauteur={hauteur}, aire={triangle.aire():.2f}")
    plt.show()


def afficher_cercle(cercle):
    r = cercle.rayon
    theta = np.linspace(0, 2*np.pi, 300)
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.plot(x, y)
    ax.fill(x, y, alpha=0.3)
    config_axes_2d(ax, f"Cercle — rayon={r}, aire={cercle.aire():.2f}")
    plt.show()


def afficher_losange(losange):
    d1 = losange.diagonale1
    d2 = losange.diagonale2
    x = [0, d1/2, 0, -d1/2, 0]
    y = [d2/2, 0, -d2/2, 0, d2/2]
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.plot(x, y)
    ax.fill(x, y, alpha=0.3)
    config_axes_2d(ax, f"Losange — aire={losange.aire():.2f}")
    plt.show()


def afficher_trapeze(trapeze):
    gb = trapeze.grande_base
    pb = trapeze.petite_base
    h = trapeze.hauteur
    decalage = (gb - pb) / 2
    x = [0, gb, gb - decalage, decalage, 0]
    y = [0, 0, h, h, 0]
    fig, ax = plt.subplots(figsize=(6, 4))
    ax.plot(x, y)
    ax.fill(x, y, alpha=0.3)
    config_axes_2d(ax, f"Trapèze — aire={trapeze.aire():.2f}")
    plt.show()


def afficher_parallelogramme(parallelogramme):
    b = parallelogramme.base
    h = parallelogramme.hauteur
    decalage = b * 0.3
    x = [0, b, b + decalage, decalage, 0]
    y = [0, 0, h, h, 0]
    fig, ax = plt.subplots(figsize=(6, 4))
    ax.plot(x, y)
    ax.fill(x, y, alpha=0.3)
    config_axes_2d(ax, f"Parallélogramme — aire={parallelogramme.aire():.2f}")
    plt.show()


def afficher_pentagone(pentagone):
    rayon = pentagone.cote / (2 * np.sin(np.pi / 5))
    x, y = polygone_regulier_points(5, rayon, rotation=np.pi/2)
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.plot(x, y)
    ax.fill(x, y, alpha=0.3)
    config_axes_2d(ax, f"Pentagone — aire={pentagone.aire():.2f}")
    plt.show()


def afficher_hexagone(hexagone):
    rayon = hexagone.cote
    x, y = polygone_regulier_points(6, rayon, rotation=np.pi/6)
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.plot(x, y)
    ax.fill(x, y, alpha=0.3)
    config_axes_2d(ax, f"Hexagone — aire={hexagone.aire():.2f}")
    plt.show()


def afficher_octogone(octogone):
    rayon = octogone.cote / (2 * np.sin(np.pi / 8))
    x, y = polygone_regulier_points(8, rayon, rotation=np.pi/8)
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.plot(x, y)
    ax.fill(x, y, alpha=0.3)
    config_axes_2d(ax, f"Octogone — aire={octogone.aire():.2f}")
    plt.show()


def afficher_ovale(ovale):
    a = ovale.demi_grand_axe
    b = ovale.demi_petit_axe
    theta = np.linspace(0, 2*np.pi, 300)
    x = a * np.cos(theta)
    y = b * np.sin(theta)
    fig, ax = plt.subplots(figsize=(6, 4))
    ax.plot(x, y)
    ax.fill(x, y, alpha=0.3)
    config_axes_2d(ax, f"Ovale — aire={ovale.aire():.2f}")
    plt.show()


def afficher_anneau(anneau):
    re = anneau.rayon_externe
    ri = anneau.rayon_interne
    theta = np.linspace(0, 2*np.pi, 300)
    xe = re * np.cos(theta)
    ye = re * np.sin(theta)
    xi = ri * np.cos(theta)
    yi = ri * np.sin(theta)
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.plot(xe, ye)
    ax.plot(xi, yi)
    ax.fill(xe, ye, alpha=0.3)
    ax.fill(xi, yi, color='white')
    config_axes_2d(ax, f"Anneau — aire={anneau.aire():.2f}")
    plt.show()


# Affichage des 12 formes 2D
afficher_carre(formes[0])
afficher_rectangle(formes[1])
afficher_triangle(formes[2])
afficher_cercle(formes[3])
afficher_losange(formes[4])
afficher_trapeze(formes[5])
afficher_parallelogramme(formes[6])
afficher_pentagone(formes[7])
afficher_hexagone(formes[8])
afficher_octogone(formes[9])
afficher_ovale(formes[10])
afficher_anneau(formes[11])
```

## Visualisation des 8 formes 3D

Cette section affiche toutes les formes 3D du rapport.


```python

import math

def config_axes_3d(ax, titre, limite):
    ax.set_title(titre)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_box_aspect([1, 1, 1])
    ax.set_xlim(-limite, limite)
    ax.set_ylim(-limite, limite)
    ax.set_zlim(0, limite if titre.startswith('Hémisphère') else limite)


def afficher_cube(cube):
    a = cube.cote
    sommets = np.array([
        [0, 0, 0], [a, 0, 0], [a, a, 0], [0, a, 0],
        [0, 0, a], [a, 0, a], [a, a, a], [0, a, a]
    ])
    faces = [
        [sommets[j] for j in [0, 1, 2, 3]],
        [sommets[j] for j in [4, 5, 6, 7]],
        [sommets[j] for j in [0, 1, 5, 4]],
        [sommets[j] for j in [2, 3, 7, 6]],
        [sommets[j] for j in [1, 2, 6, 5]],
        [sommets[j] for j in [4, 7, 3, 0]],
    ]
    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.add_collection3d(Poly3DCollection(faces, alpha=0.3))
    ax.scatter(sommets[:, 0], sommets[:, 1], sommets[:, 2])
    ax.set_xlim(0, a)
    ax.set_ylim(0, a)
    ax.set_zlim(0, a)
    ax.set_box_aspect([1, 1, 1])
    ax.set_title(f"Cube — côté={a}, aire={cube.aire():.2f}, volume={cube.volume():.2f}")
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.show()


def afficher_sphere(sphere):
    r = sphere.rayon
    u = np.linspace(0, 2 * np.pi, 60)
    v = np.linspace(0, np.pi, 60)
    x = r * np.outer(np.cos(u), np.sin(v))
    y = r * np.outer(np.sin(u), np.sin(v))
    z = r * np.outer(np.ones(np.size(u)), np.cos(v))
    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, alpha=0.3)
    ax.set_xlim(-r, r)
    ax.set_ylim(-r, r)
    ax.set_zlim(-r, r)
    ax.set_box_aspect([1, 1, 1])
    ax.set_title(f"Sphère — rayon={r}, aire={sphere.aire():.2f}, volume={sphere.volume():.2f}")
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.show()


def afficher_cylindre(cylindre):
    r = cylindre.rayon
    h = cylindre.hauteur
    z = np.linspace(0, h, 50)
    theta = np.linspace(0, 2*np.pi, 50)
    theta_grid, z_grid = np.meshgrid(theta, z)
    x_grid = r * np.cos(theta_grid)
    y_grid = r * np.sin(theta_grid)
    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x_grid, y_grid, z_grid, alpha=0.3)
    ax.set_title(f"Cylindre — rayon={r}, hauteur={h}, aire={cylindre.aire():.2f}, volume={cylindre.volume():.2f}")
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.show()


def afficher_cone(cone):
    r = cone.rayon
    h = cone.hauteur
    theta = np.linspace(0, 2*np.pi, 50)
    t = np.linspace(0, 1, 50)
    theta_grid, t_grid = np.meshgrid(theta, t)
    x = (1 - t_grid) * r * np.cos(theta_grid)
    y = (1 - t_grid) * r * np.sin(theta_grid)
    z = t_grid * h
    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, alpha=0.3)
    ax.set_title(f"Cône — aire={cone.aire():.2f}, volume={cone.volume():.2f}")
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.show()


def afficher_pyramide(pyramide):
    cote = math.sqrt(pyramide.aire_base)
    h = pyramide.hauteur
    base = np.array([
        [-cote/2, -cote/2, 0], [cote/2, -cote/2, 0],
        [cote/2, cote/2, 0], [-cote/2, cote/2, 0]
    ])
    apex = np.array([0, 0, h])
    faces = [
        [base[0], base[1], base[2], base[3]],
        [base[0], base[1], apex],
        [base[1], base[2], apex],
        [base[2], base[3], apex],
        [base[3], base[0], apex],
    ]
    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.add_collection3d(Poly3DCollection(faces, alpha=0.3))
    ax.set_xlim(-cote, cote)
    ax.set_ylim(-cote, cote)
    ax.set_zlim(0, h)
    ax.set_box_aspect([1, 1, 1])
    ax.set_title(f"Pyramide — volume={pyramide.volume():.2f}")
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.show()


def afficher_prisme(prisme):
    cote = 2
    hauteur_triangle = np.sqrt(3)
    base1 = np.array([
        [-cote, 0, 0], [cote, 0, 0], [0, hauteur_triangle, 0]
    ])
    base2 = base1 + np.array([0, 0, prisme.hauteur])
    faces = [
        [base1[0], base1[1], base1[2]],
        [base2[0], base2[1], base2[2]],
        [base1[0], base1[1], base2[1], base2[0]],
        [base1[1], base1[2], base2[2], base2[1]],
        [base1[2], base1[0], base2[0], base2[2]],
    ]
    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.add_collection3d(Poly3DCollection(faces, alpha=0.3))
    ax.set_xlim(-3, 3)
    ax.set_ylim(-1, 4)
    ax.set_zlim(0, prisme.hauteur)
    ax.set_box_aspect([1, 1, 1])
    ax.set_title(f"Prisme — volume={prisme.volume():.2f}")
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.show()


def afficher_tore(tore):
    R = tore.grand_rayon
    r = tore.petit_rayon
    u = np.linspace(0, 2*np.pi, 60)
    v = np.linspace(0, 2*np.pi, 60)
    u, v = np.meshgrid(u, v)
    x = (R + r*np.cos(v)) * np.cos(u)
    y = (R + r*np.cos(v)) * np.sin(u)
    z = r * np.sin(v)
    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, alpha=0.3)
    ax.set_title(f"Tore — aire={tore.aire():.2f}, volume={tore.volume():.2f}")
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.show()


def afficher_hemisphere(hemisphere):
    r = hemisphere.rayon
    u = np.linspace(0, 2*np.pi, 60)
    v = np.linspace(0, np.pi/2, 60)
    x = r * np.outer(np.cos(u), np.sin(v))
    y = r * np.outer(np.sin(u), np.sin(v))
    z = r * np.outer(np.ones(np.size(u)), np.cos(v))
    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, alpha=0.3)
    ax.set_xlim(-r, r)
    ax.set_ylim(-r, r)
    ax.set_zlim(0, r)
    ax.set_box_aspect([1, 1, 0.7])
    ax.set_title(f"Hémisphère — aire={hemisphere.aire():.2f}, volume={hemisphere.volume():.2f}")
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.show()


# Affichage des 8 formes 3D
afficher_cube(formes[12])
afficher_sphere(formes[13])
afficher_cylindre(formes[14])
afficher_cone(formes[15])
afficher_pyramide(formes[16])
afficher_prisme(formes[17])
afficher_tore(formes[18])
afficher_hemisphere(formes[19])
```
