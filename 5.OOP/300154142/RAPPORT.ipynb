```python
from Carre import Carre
from Cercle import Cercle
from Triangle import Triangle
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
c1 = Carre(4)
c2 = Cercle(3)
t1 = Triangle(5, 2)
```

```python
print(c1.afficher_info(), c1.aire())
print(c2.afficher_info(), round(c2.aire(), 2))
print(t1.afficher_info(), t1.aire())
```

```python
def afficher_carre(carre):
    cote = carre.cote
    x = [0, cote, cote, 0, 0]
    y = [0, 0, cote, cote, 0]

    plt.figure()
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Carré - aire = {carre.aire()}")
    plt.axis("equal")
    plt.grid()
    plt.show()
```

```python
def afficher_cercle(cercle):
    r = cercle.rayon
    theta = np.linspace(0, 2*np.pi, 100)

    x = r * np.cos(theta)
    y = r * np.sin(theta)

    plt.figure()
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Cercle - aire = {round(cercle.aire(), 2)}")
    plt.axis("equal")
    plt.grid()
    plt.show()
```

```python
def afficher_triangle(triangle):
    base = triangle.base
    hauteur = triangle.hauteur

    x = [0, base, 0, 0]
    y = [0, 0, hauteur, 0]

    plt.figure()
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Triangle - aire = {triangle.aire()}")
    plt.axis("equal")
    plt.grid()
    plt.show()
```

```python
afficher_carre(c1)
afficher_cercle(c2)
afficher_triangle(t1)
```
