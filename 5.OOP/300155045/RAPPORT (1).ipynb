## ▶️ 1. Importer les classes

```python
from Carre import Carre
from Cercle import Cercle
from Triangle import Triangle

c1 = Carre(4)
c2 = Cercle(3)
c3 = Triangle(5, 2)

formes = [c1, c2, c3]
for f in formes:
    print(f.afficher_info())
```

## 🎨 2. Imports pour l'affichage graphique

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 🔲 3. Tracer un Carré

```python
def afficher_carre(carre):
    cote = carre.cote
    x = [0, cote, cote, 0, 0]
    y = [0, 0, cote, cote, 0]
    plt.figure(figsize=(5, 5))
    plt.plot(x, y, 'b-', linewidth=2)
    plt.fill(x, y, alpha=0.3, color='blue')
    plt.title(f"Carré — côté={cote}, aire={carre.aire()}")
    plt.axis('equal')
    plt.grid(True)
    plt.show()

afficher_carre(c1)
```

## ⭕ 4. Tracer un Cercle

```python
def afficher_cercle(cercle):
    r = cercle.rayon
    theta = np.linspace(0, 2*np.pi, 300)
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    plt.figure(figsize=(5, 5))
    plt.plot(x, y, 'r-', linewidth=2)
    plt.fill(x, y, alpha=0.3, color='red')
    plt.title(f"Cercle — rayon={r}, aire={cercle.aire():.2f}")
    plt.axis('equal')
    plt.grid(True)
    plt.show()

afficher_cercle(c2)
```

## 🔺 5. Tracer un Triangle

```python
def afficher_triangle(triangle):
    b = triangle.base
    h = triangle.hauteur
    x = [0, b, b/2, 0]
    y = [0, 0, h, 0]
    plt.figure(figsize=(5, 5))
    plt.plot(x, y, 'g-', linewidth=2)
    plt.fill(x, y, alpha=0.3, color='green')
    plt.title(f"Triangle — base={b}, hauteur={h}, aire={triangle.aire()}")
    plt.axis('equal')
    plt.grid(True)
    plt.show()

afficher_triangle(c3)
```

## 📊 6. Comparaison des aires

```python
noms = [f.nom for f in formes]
aires = [f.aire() for f in formes]

plt.figure(figsize=(8, 5))
bars = plt.bar(noms, aires, color=['blue', 'red', 'green'])
plt.title('Comparaison des Aires')
plt.xlabel('Figure')
plt.ylabel('Aire')
for bar, aire in zip(bars, aires):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.2,
             f'{aire:.2f}', ha='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.grid(axis='y', alpha=0.5)
plt.show()
```
