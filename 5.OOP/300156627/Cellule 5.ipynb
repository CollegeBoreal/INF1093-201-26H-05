```python
class Carre:
    def __init__(self, cote):
        self.cote = cote

    def afficher_info(self):
        return f"Carré de côté {self.cote}"

class Cercle:
    def __init__(self, rayon):
        self.rayon = rayon

    def afficher_info(self):
        return f"Cercle de rayon {self.rayon}"

class Triangle:
    def __init__(self, base, hauteur):
        self.base = base
        self.hauteur = hauteur

    def afficher_info(self):
        return f"Triangle de base {self.base} et hauteur {self.hauteur}"
formes = [Carre(4), Cercle(3), Triangle(5, 2)]

for f in formes:
    print(f.afficher_info())
```
