```python
from Carre import Carre
from Cercle import Cercle
from Triangle import Triangle
from Rectangle import Rectangle
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def afficher_carre(carre):
    cote = carre.cote
    x = [0, cote, cote, 0, 0]
    y = [0, 0, cote, cote, 0]

    plt.figure(figsize=(5,5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Carré — côté={cote}, aire={carre.aire()}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()


def afficher_cercle(cercle):
    r = cercle.rayon
    theta = np.linspace(0, 2*np.pi, 300)
    x = r * np.cos(theta)
    y = r * np.sin(theta)

    plt.figure(figsize=(5,5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Cercle — rayon={r}, aire={cercle.aire():.2f}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()


def afficher_triangle(triangle):
    base = triangle.base
    hauteur = triangle.hauteur

    x = [0, base, base/2, 0]
    y = [0, 0, hauteur, 0]

    plt.figure(figsize=(5,5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Triangle — base={base}, hauteur={hauteur}, aire={triangle.aire()}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()


def afficher_rectangle(rect):
    L = rect.longueur
    l = rect.largeur

    x = [0, L, L, 0, 0]
    y = [0, 0, l, l, 0]

    plt.figure(figsize=(5,5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Rectangle — L={L}, l={l}, aire={rect.aire()}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()

def afficher_pentagone(p):
    n = 5
    r = p.cote

    angles = np.linspace(0, 2*np.pi, n+1)
    x = r * np.cos(angles)
    y = r * np.sin(angles)

    plt.figure(figsize=(5,5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Pentagone — côté={p.cote}, aire={p.aire():.2f}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()

def afficher_ellipse(e):
    a = e.a
    b = e.b

    theta = np.linspace(0, 2*np.pi, 300)
    x = a * np.cos(theta)
    y = b * np.sin(theta)

    plt.figure(figsize=(5,5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Ellipse — a={a}, b={b}, aire={e.aire():.2f}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()

def afficher_hexagone(h):
    n = 6
    r = h.cote

    angles = np.linspace(0, 2*np.pi, n+1)
    x = r * np.cos(angles)
    y = r * np.sin(angles)

    plt.figure(figsize=(5,5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Hexagone — côté={h.cote}, aire={h.aire():.2f}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()

def afficher_losange(l):
    d1 = l.d1
    d2 = l.d2

    x = [0, d1/2, 0, -d1/2, 0]
    y = [d2/2, 0, -d2/2, 0, d2/2]

    plt.figure(figsize=(5,5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Losange — aire={l.aire()}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()

def afficher_segment(longueur):
    x = [0, longueur]
    y = [0, 0]

    plt.figure()
    plt.plot(x, y)
    plt.title(f"Segment — longueur={longueur}")
    plt.grid(True)
    plt.show()
```
