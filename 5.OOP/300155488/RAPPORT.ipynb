```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import math
import os

# ===============================
# DOSSIER
# ===============================
os.makedirs("images", exist_ok=True)

# ===============================
# CLASSES
# ===============================
class Figure:
    def __init__(self, nom):
        self.nom = nom
    def aire(self):
        raise NotImplementedError

class Carre(Figure):
    def __init__(self, cote):
        super().__init__("Carré")
        self.cote = cote
    def aire(self):
        return self.cote ** 2

class Cercle(Figure):
    def __init__(self, rayon):
        super().__init__("Cercle")
        self.rayon = rayon
    def aire(self):
        return math.pi * self.rayon ** 2

class Triangle(Figure):
    def __init__(self, base, hauteur):
        super().__init__("Triangle")
        self.base = base
        self.hauteur = hauteur
    def aire(self):
        return (self.base * self.hauteur) / 2

# ===============================
# DONNÉES
# ===============================
carre = Carre(4)
cercle = Cercle(3)
triangle = Triangle(5, 2)

c = carre.cote
r = cercle.rayon
b = triangle.base
h = triangle.hauteur

# =========================================================
# 2D — CHAQUE FIGURE SEULE
# =========================================================

# ---- Carré
plt.figure(figsize=(6,6))
x = [0,c,c,0,0]
y = [0,0,c,c,0]
plt.plot(x,y,linewidth=3)
plt.fill(x,y,alpha=0.3)
plt.text(c/2,-0.3,f"{c}",ha="center")
plt.text(-0.3,c/2,f"{c}",va="center",rotation=90)
plt.title(f"Carré | Aire={carre.aire():.2f}")
plt.gca().set_aspect("equal")
plt.grid()
plt.savefig("images/carre.png", dpi=300)
plt.show()

# ---- Cercle
plt.figure(figsize=(6,6))
theta = np.linspace(0,2*np.pi,400)
x = r*np.cos(theta)
y = r*np.sin(theta)
plt.plot(x,y,linewidth=3)
plt.fill(x,y,alpha=0.3)
plt.plot([0,r],[0,0],linewidth=2)
plt.text(r/2,0.2,f"r={r}")
plt.title(f"Cercle | Aire={cercle.aire():.2f}")
plt.gca().set_aspect("equal")
plt.grid()
plt.savefig("images/cercle.png", dpi=300)
plt.show()

# ---- Triangle
plt.figure(figsize=(6,6))
x = [0,b,b/2,0]
y = [0,0,h,0]
plt.plot(x,y,linewidth=3)
plt.fill(x,y,alpha=0.3)
plt.text(b/2,-0.3,f"b={b}",ha="center")
plt.plot([b/2,b/2],[0,h],"--")
plt.text(b/2+0.2,h/2,f"h={h}")
plt.title(f"Triangle | Aire={triangle.aire():.2f}")
plt.gca().set_aspect("equal")
plt.grid()
plt.savefig("images/triangle.png", dpi=300)
plt.show()

# =========================================================
# 3D — CHAQUE FIGURE SEULE
# =========================================================

# ---- Carré 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

pts = [[0,0,0],[c,0,0],[c,c,0],[0,c,0]]
ax.add_collection3d(Poly3DCollection([pts], alpha=0.5))
ax.text(c/2, -0.5, 0, f"{c}")
ax.set_title("Carré 3D")
plt.savefig("images/carre_3d.png", dpi=300)
plt.show()

# ---- Cercle 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

theta = np.linspace(0,2*np.pi,200)
x = r*np.cos(theta)
y = r*np.sin(theta)
z = np.zeros_like(theta)

ax.plot(x,y,z,linewidth=3)
ax.plot([0,r],[0,0],[0,0])
ax.text(r/2,0,0,f"r={r}")
ax.set_title("Cercle 3D")
plt.savefig("images/cercle_3d.png", dpi=300)
plt.show()

# ---- Triangle 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

pts = [[0,0,0],[b,0,0],[b/2,h,0]]
ax.add_collection3d(Poly3DCollection([pts], alpha=0.6))
ax.text(b/2,-0.5,0,f"b={b}")
ax.text(b/2,h/2,0,f"h={h}")
ax.set_title("Triangle 3D")
plt.savefig("images/triangle_3d.png", dpi=300)
plt.show()


plt.show()
```
