```python
import matplotlib.pyplot as plt
import numpy as np

theta = np.linspace(0, 2*np.pi, 100)
x = 3 * np.cos(theta)
y = 3 * np.sin(theta)

plt.plot(x, y)
plt.title("Cercle")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()

# 🔵 Cercle
theta = np.linspace(0, 2*np.pi, 200)
x_c = 3 * np.cos(theta)
y_c = 3 * np.sin(theta)
plt.plot(x_c, y_c)
plt.fill(x_c, y_c, alpha=0.2)

# 🟥 Carré
x_s = [-2, 2, 2, -2, -2]
y_s = [-2, -2, 2, 2, -2]
plt.plot(x_s, y_s)
plt.fill(x_s, y_s, alpha=0.2)

# 🔺 Triangle
x_t = [0, -2, 2, 0]
y_t = [3, -2, -2, 3]
plt.plot(x_t, y_t)
plt.fill(x_t, y_t, alpha=0.2)

# 🎯 Mise en forme
plt.title("Composition de figures géométriques")
plt.axis("equal")
plt.grid(True)

plt.show()
```

```python
theta = np.linspace(0, 6*np.pi, 500)
r = theta

x = r * np.cos(theta)
y = r * np.sin(theta)

plt.plot(x, y)
plt.title("Spirale")
plt.grid(True)
plt.show()
```

```python
theta = np.linspace(0, 2*np.pi, 500)
r = np.sin(5 * theta)

x = r * np.cos(theta)
y = r * np.sin(theta)

plt.plot(x, y)
plt.title("Fleur mathématique")
plt.axis("equal")
plt.grid(True)
plt.show()
```
