Ce notebook démontre l'utilisation d'une hiérarchie de classes Python :
- `Figure` (classe de base)

Bibliothèques utilisées :
- **matplotlib 3.9.2**
- **numpy 2.1.3**

```python
from Carre import Carre
from Cercle import Cercle

c1 = Carre(4)
c2 = Cercle(3)

print(c1.afficher_info())
print(c2.afficher_info())
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 🟦 Visualisation du Carré

```python
def afficher_carre(carre):
    cote = carre.cote
    x = [0, cote, cote, 0, 0]
    y = [0, 0, cote, cote, 0]
    plt.figure(figsize=(5, 5))
    plt.plot(x, y, 'b-')
    plt.fill(x, y, alpha=0.3, color='blue')
    plt.title(f'Carré — côté={cote}, aire={carre.aire()}')
    plt.axis('equal')
    plt.grid(True)
    plt.savefig('images/carre.png')
    plt.show()

afficher_carre(c1)
```

## ⚪ Visualisation du Cercle

```python
def afficher_cercle(cercle):
    r = cercle.rayon
    theta = np.linspace(0, 2*np.pi, 300)
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    plt.figure(figsize=(5, 5))
    plt.plot(x, y, 'r-')
    plt.fill(x, y, alpha=0.3, color='red')
    plt.title(f'Cercle — rayon={r}, aire={cercle.aire():.2f}')
    plt.axis('equal')
    plt.grid(True)
    plt.savefig('images/cercle.png')
    plt.show()

afficher_cercle(c2)
```

## 🔺 Visualisation du Triangle

```python
from Triangle import Triangle

t1 = Triangle(5, 2)
print(t1.afficher_info())

def afficher_triangle(triangle):
    b = triangle.base
    h = triangle.hauteur
    x = [0, b, b/2, 0]
    y = [0, 0, h, 0]
    plt.figure(figsize=(5, 5))
    plt.plot(x, y, 'g-')
    plt.fill(x, y, alpha=0.3, color='green')
    plt.title(f'Triangle — base={b}, hauteur={h}, aire={triangle.aire()}')
    plt.axis('equal')
    plt.grid(True)
    plt.savefig('images/triangle.png')
    plt.show()

afficher_triangle(t1)
```
