```python
import matplotlib.pyplot as plt

noms = []
notes = []

with open("etudiants.txt", "r", encoding="utf-8") as f:
    for ligne in f:
        nom, note = ligne.strip().split(",")
        noms.append(nom)
        notes.append(float(note))

plt.figure(figsize=(10,5))
plt.bar(noms, notes, color='skyblue')
plt.axhline(sum(notes)/len(notes), color='red', linestyle='--', label='Moyenne')
plt.title("Notes des étudiant.e.s")
plt.xlabel("Étudiant.e.s")
plt.ylabel("Notes")
plt.legend()
plt.show()
```
