```python
import networkx as nx
import matplotlib.pyplot as plt

# Créer le graphe
G = nx.Graph()

# Ajouter les arêtes
edges = [
    ('a', 'b', 7), ('a', 'c', 9), ('a', 'f', 14),
    ('b', 'c', 10), ('b', 'd', 15),
    ('c', 'd', 11), ('c', 'f', 2),
    ('d', 'e', 6),
    ('e', 'f', 9)
]

for u, v, w in edges:
    G.add_edge(u, v, weight=w)

# Chemin le plus court (résultat de Dijkstra)
path = ['a', 'c', 'f', 'e']

# Positions des nœuds
pos = {
    'a': (0, 2), 'b': (2, 3), 'c': (2, 1),
    'd': (4, 2), 'e': (6, 2), 'f': (4, 0)
}

# Visualisation
plt.figure(figsize=(12, 8))

# Nœuds
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=700)
nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold')

# Arêtes normales (gris)
nx.draw_networkx_edges(G, pos, width=1, alpha=0.5, edge_color='gray')

# Chemin le plus court (rouge)
path_edges = list(zip(path, path[1:]))
nx.draw_networkx_edges(G, pos, edgelist=path_edges, width=4, edge_color='red')

# Labels des poids
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=10)

# Départ et arrivée
nx.draw_networkx_nodes(G, pos, nodelist=['a'], node_color='green', node_size=800, label='Départ (a)')
nx.draw_networkx_nodes(G, pos, nodelist=['e'], node_color='orange', node_size=800, label='Arrivée (e)')

plt.title("Plus court chemin de 'a' à 'e': a → c → f → e (distance = 20)", fontsize=14)
plt.axis('off')
plt.legend()
plt.tight_layout()
plt.show()

print("✅ Chemin trouvé: a → c → f → e")
print("📏 Distance totale: 20")
```
