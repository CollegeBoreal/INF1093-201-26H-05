```python
# RAPPORT DIJKSTRA FINAL - AUCUNE ERREUR

from graph import Graph
from dijkstra_tp import g, dijkstra, shortest

print("🏆 TP DIJKSTRA COLLÈGE BORÉAL")
print("SOMMETS:", list(g.get_vertices()))

import networkx as nx
import matplotlib.pyplot as plt

# CONSTRUCTION GRAPHE
G_nx = nx.Graph()
for v in g: G_nx.add_node(v.get_id())
for v in g:
    for w in v.get_connections():
        G_nx.add_edge(v.get_id(), w.get_id(), weight=v.get_weight(w))

pos = nx.spring_layout(G_nx, seed=42)

# G R A P H I Q U E   1
plt.figure(figsize=(10,7))
nx.draw(G_nx, pos, with_labels=True, node_color='#ADD8E6', 
        node_size=1400, font_size=14, font_weight='bold')
nx.draw_networkx_edge_labels(G_nx, pos, 
    edge_labels=nx.get_edge_attributes(G_nx,'weight'))
plt.title("GRAPHE ORIENTÉ PONDERÉ", fontsize=18)
plt.axis('off')
plt.savefig('images/graphe_complet.png', dpi=300, bbox_inches='tight')
plt.show()

# DIJKSTRA
start = g.get_vertex('a')
target = g.get_vertex('e')
dijkstra(g, start)
path = shortest(target)
print(f"\n✅ CHEMIN a → e: {path} (distance: {target.get_distance()})")

# G R A P H I Q U E   2
plt.figure(figsize=(12,8))
nx.draw(G_nx, pos, node_color='#ADD8E6', node_size=1100, 
        with_labels=True, font_size=13, font_weight='bold')
nx.draw_networkx_edge_labels(G_nx, pos, 
    edge_labels=nx.get_edge_attributes(G_nx,'weight'))

path_edges = list(zip(path, path[1:]))
nx.draw_networkx_nodes(G_nx, pos, nodelist=path, node_color='#FFD700', node_size=2000)
nx.draw_networkx_edges(G_nx, pos, edgelist=path_edges, edge_color='red', width=8)

plt.title(f"PLUS COURT CHEMIN\n{' → '.join(path)}\nDistance = {target.get_distance()}", 
          fontsize=20, pad=30)
plt.axis('off')
plt.savefig('images/chemin_optimal.png', dpi=300, bbox_inches='tight')
plt.show()

# VÉRIFICATION
expected_path = ['a', 'c', 'f', 'e']
print("\n" + "═"*60)
print("VERIFICATION:")
print(f"Calculé : {path}")
print(f"Attendu  : {expected_path}")
print("✅ 100% CORRECT !" if path == expected_path else "❌ Erreur")

print("\nDISTANCES:")
for v in g:
    print(f"{v.get_id()}: {v.get_distance()}")

print("\n💾 PNG sauvés: graphe_complet.png + chemin_optimal.png")
print("🏆 TP TERMINÉ!")
```
