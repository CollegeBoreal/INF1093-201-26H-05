## 📌 Objectif
Trouver le plus court chemin entre des capitales européennes avec l'algorithme de Dijkstra.

```python
from graph import Graph
from dijkstra_tp import g, dijkstra, shortest
```

## 🗺️ Graphe des capitales

```python
print('=== Connexions du graphe ===')
for v in g:
    for w in v.get_connections():
        print(f'  {v.get_id():12s} <--> {w.get_id():12s}  {v.get_weight(w):5d} km')
```

## 📍 Plus court chemin : Paris → Prague

```python
target = g.get_vertex('Prague')
path = shortest(target)
print('Chemin :', ' → '.join(path))
print('Distance :', target.get_distance(), 'km')
```
