```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

## 1. Construction du graphe

```python
# Création du graphe avec networkx
G = nx.Graph()

# Ajout des noeuds
nodes = ['a', 'b', 'c', 'd', 'e', 'f']
G.add_nodes_from(nodes)

# Ajout des arêtes avec poids
edges = [
    ('a', 'b', 7),
    ('a', 'c', 9),
    ('a', 'f', 14),
    ('b', 'c', 10),
    ('b', 'd', 15),
    ('c', 'd', 11),
    ('c', 'f', 2),
    ('d', 'e', 6),
    ('e', 'f', 9),
]
G.add_weighted_edges_from(edges)

print("Noeuds:", list(G.nodes()))
print("Arêtes:", list(G.edges(data=True)))
```

## 2. Visualisation du graphe

```python
# Positions fixes pour un beau rendu
pos = {
    'a': (0, 1),
    'b': (1, 2),
    'c': (1, 0),
    'd': (2.5, 2),
    'e': (3.5, 1),
    'f': (2, 0.5),
}

# Chemin le plus court trouvé par Dijkstra
shortest_path = ['a', 'c', 'f', 'e']
path_edges = list(zip(shortest_path, shortest_path[1:]))

# Couleurs des noeuds
node_colors = []
for node in G.nodes():
    if node == 'a':
        node_colors.append('#2ecc71')  # vert = départ
    elif node == 'e':
        node_colors.append('#e74c3c')  # rouge = arrivée
    elif node in shortest_path:
        node_colors.append('#f39c12')  # orange = sur le chemin
    else:
        node_colors.append('#3498db')  # bleu = autres

# Couleurs des arêtes
edge_colors = []
edge_widths = []
for u, v in G.edges():
    if (u, v) in path_edges or (v, u) in path_edges:
        edge_colors.append('#e74c3c')  # rouge = chemin optimal
        edge_widths.append(4)
    else:
        edge_colors.append('#bdc3c7')  # gris = autres
        edge_widths.append(1.5)

# Poids des arêtes
edge_labels = nx.get_edge_attributes(G, 'weight')

# Affichage
plt.figure(figsize=(10, 6))
plt.title("Graphe de Dijkstra\nChouaib Ait Chouaib | 300155045", fontsize=14, fontweight='bold')

nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=800)
nx.draw_networkx_labels(G, pos, font_size=14, font_weight='bold', font_color='white')
nx.draw_networkx_edges(G, pos, edge_color=edge_colors, width=edge_widths)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=11, font_color='#2c3e50')

# Légende
legend_handles = [
    mpatches.Patch(color='#2ecc71', label='Départ (a)'),
    mpatches.Patch(color='#e74c3c', label='Arrivée (e)'),
    mpatches.Patch(color='#f39c12', label='Chemin optimal'),
    mpatches.Patch(color='#3498db', label='Autres noeuds'),
    mpatches.Patch(color='#e74c3c', label='Plus court chemin (a→c→f→e)'),
]
plt.legend(handles=legend_handles, loc='upper right', fontsize=9)

plt.axis('off')
plt.tight_layout()
plt.savefig('graphe_dijkstra.png', dpi=150, bbox_inches='tight')
plt.show()
print("Graphe sauvegardé: graphe_dijkstra.png")
```

## 3. Exécution de l'algorithme de Dijkstra

```python
# Utilisation de l'implémentation networkx pour vérification
path_nx = nx.dijkstra_path(G, 'a', 'e', weight='weight')
length_nx = nx.dijkstra_path_length(G, 'a', 'e', weight='weight')

print("=== Résultats Dijkstra ===")
print(f"Plus court chemin (a → e) : {' → '.join(path_nx)}")
print(f"Distance totale           : {length_nx}")
print()

# Détail des distances depuis 'a'
lengths = nx.single_source_dijkstra_path_length(G, 'a', weight='weight')
print("Distance depuis 'a' vers tous les noeuds :")
for node in sorted(lengths):
    print(f"  a → {node} : {lengths[node]}")
```

## 4. Vérification du résultat

```python
expected_path = ['a', 'c', 'f', 'e']
expected_distance = 20

print("=== Auto-correction ===")
if path_nx == expected_path and length_nx == expected_distance:
    print("✅ Bravo, le chemin est correct !")
    print(f"   Chemin trouvé   : {' → '.join(path_nx)}")
    print(f"   Distance totale : {length_nx}")
else:
    print("❌ Chemin incorrect.")
    print(f"   Chemin obtenu   : {path_nx}")
    print(f"   Chemin attendu  : {expected_path}")
```

## 5. Conclusion

L'algorithme de Dijkstra trouve le chemin le plus court entre le noeud **a** et le noeud **e** :

**a → c → f → e** avec une distance totale de **20**.

Détail du chemin :
- a → c : poids 9
- c → f : poids 2
- f → e : poids 9
- **Total : 9 + 2 + 9 = 20**
