```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()

edges = [
    ("a", "b", 7),
    ("a", "c", 9),
    ("a", "f", 14),
    ("b", "c", 10),
    ("b", "d", 15),
    ("c", "d", 11),
    ("c", "f", 2),
    ("d", "e", 6),
    ("e", "f", 9),
]

for u, v, w in edges:
    G.add_edge(u, v, weight=w)

pos = nx.spring_layout(G, seed=42)

# chemin en rouge
chemin = [("a", "c"), ("c", "f"), ("f", "e")]

nx.draw(G, pos, with_labels=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u,v):w for u,v,w in edges})
nx.draw_networkx_edges(G, pos, edgelist=chemin, edge_color="red", width=3)

plt.title("Graphe Dijkstra")
plt.show()
```
