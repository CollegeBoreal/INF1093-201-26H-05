## Graphe utilisé\nSommets : A, B, C, D, E, F, G\n\nArêtes pondérées :\n- A-B (4)\n- A-C (3)\n- B-D (5)\n- B-E (12)\n- C-D (8)\n- C-F (7)\n- D-E (2)\n- D-F (6)\n- E-G (3)\n- F-G (4)

```python
import networkx as nx\nimport matplotlib.pyplot as plt\n\nG = nx.Graph()\nedges = [\n    ('A','B',4), ('A','C',3), ('B','D',5), ('B','E',12),\n    ('C','D',8), ('C','F',7), ('D','E',2), ('D','F',6),\n    ('E','G',3), ('F','G',4)\n]\n\nfor u, v, w in edges:\n    G.add_edge(u, v, weight=w)\n\npos = {\n    'A': (0, 1), 'B': (1, 2), 'C': (1, 0), 'D': (2, 1),\n    'E': (3, 2), 'F': (3, 0), 'G': (4, 1)\n}\n\nshortest_path = ['A', 'C', 'F', 'G']\npath_edges = list(zip(shortest_path, shortest_path[1:]))\n\nplt.figure(figsize=(8, 5))\nnx.draw(G, pos, with_labels=True, node_size=2000)\nnx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G, 'weight'))\nnx.draw_networkx_edges(G, pos, edgelist=path_edges, width=3)\nplt.title('Visualisation du graphe et du plus court chemin')\nplt.show()
```

## Conclusion\nLe plus court chemin trouvé entre A et G dans cette version du graphe est **A → C → F → G**, avec une distance totale de **14**.
