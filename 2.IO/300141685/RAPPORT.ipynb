```python
import matplotlib.pyplot as plt

noms = []
notes = []

with open("etudiants.txt", "r", encoding="utf-8") as f:
    for ligne in f:
        ligne = ligne.strip()
        if ligne:
            if ligne.startswith('\ufeff'):
                ligne = ligne[1:]
            parties = ligne.split(",")
            if len(parties) == 2:
                noms.append(parties[0])
                notes.append(float(parties[1]))

print("Liste des etudiants:")
for i in range(len(noms)):
    print(f"  {noms[i]} : {notes[i]}")

moyenne = sum(notes) / len(notes)
print(f"\nMoyenne du groupe : {moyenne:.2f}")
```

```python
plt.figure(figsize=(8,5))
plt.bar(noms, notes, color='skyblue', edgecolor='black')
plt.axhline(y=moyenne, color='red', linestyle='--', label=f'Moyenne = {moyenne:.2f}')
plt.title("Distribution des notes par étudiant")
plt.xlabel("Étudiants")
plt.ylabel("Notes")
plt.ylim(0, 100)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

import os
os.makedirs("images", exist_ok=True)
plt.savefig("images/diagramme_notes.png", dpi=150, bbox_inches='tight')
plt.show()

print("✅ Figure sauvegardée dans images/diagramme_notes.png")
```
