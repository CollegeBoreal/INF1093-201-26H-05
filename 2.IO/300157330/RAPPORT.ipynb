```python
Amine 68
Lina 75
Sofiane 59
Nour 82
Yasmine 73
noms = []
notes = []

with open("groupeB.txt", "r") as f:
    for ligne in f:
        nom, note = ligne.split()
        noms.append(nom)
        notes.append(int(note))
```

```python
moyenne = sum(notes) / len(notes)
moyenne
round(moyenne, 2)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

plt.bar(noms, notes)
plt.axhline(moyenne, linestyle='--',
            label=f"Moyenne : {round(moyenne, 2)}")

plt.xlabel("Ã‰tudiant.e.s")
plt.ylabel("Notes")
plt.title("Notes du groupe B avec moyenne")

plt.xticks(rotation=45, ha="right")
plt.legend()
plt.tight_layout()

plt.show()
```
