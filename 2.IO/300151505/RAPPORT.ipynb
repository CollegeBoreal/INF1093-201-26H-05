## Objectifs
- Lire un fichier texte
- Traiter les données avec Python
- Calculer la moyenne
- Générer un fichier de sortie
- Afficher un diagramme des notes

```python
noms = []
notes = []

with open("etudiants.txt", "r", encoding="utf-8") as f:
    for ligne in f:
        ligne = ligne.strip()
        if ligne:
            nom, note = ligne.split()
            noms.append(nom)
            notes.append(float(note))

print("Noms :", noms)
print("Notes :", notes)
print("Moyenne :", sum(notes) / len(notes))
```

```python
import matplotlib.pyplot as plt

# Données
noms = ["walid", "Bob", "nassim","djalloul"]
notes = [62, 76, 99,80]

# Calcul de la moyenne
moyenne = sum(notes) / len(notes)

# Affichage de la moyenne dans la console
print(f"Moyenne du groupe: {moyenne:.2f}")

# Graphique
plt.figure(figsize=(8, 5))
plt.bar(noms, notes, color="skyblue")
plt.axhline(y=moyenne, color="red", linestyle="--", label=f"Moyenne: {moyenne:.2f}")

plt.title("Notes des étudiant.e.s avec moyenne")
plt.xlabel("Étudiant.e.s")
plt.ylabel("Notes")
plt.legend()

plt.show()
```
