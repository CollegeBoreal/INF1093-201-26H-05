```python
import matplotlib.pyplot as plt

noms = []
notes = []

with open("etudiants.txt", "r", encoding="utf-8") as f:
    for ligne in f:
        ligne = ligne.strip()

        # Ignorer les lignes vides
        if ligne == "":
            continue

        parties = ligne.split()

        # Vérifier le format
        if len(parties) != 2:
            continue

        nom = parties[0]
        note = parties[1]

        noms.append(nom)
        notes.append(float(note))

# Calcul de la moyenne
moyenne = sum(notes) / len(notes)

# Graphique
plt.bar(noms, notes)
plt.axhline(y=moyenne, linestyle='--', label="Moyenne : " + str(round(moyenne, 2)))
plt.title("Notes des étudiant.e.s avec moyenne")
plt.xlabel("Étudiant.e.s")
plt.ylabel("Notes")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
