Ce notebook montre la lecture du fichier `etudiants.txt`, le calcul de la moyenne, puis l'affichage d'un diagramme des notes.

```python
from pathlib import Path

fichier = Path('etudiants.txt')
noms = []
notes = []

with fichier.open('r', encoding='utf-8') as f:
    for ligne in f:
        ligne = ligne.strip()
        if not ligne:
            continue
        parties = ligne.split()
        nom = ' '.join(parties[:-1])
        note = float(parties[-1])
        noms.append(nom)
        notes.append(note)

print('Nombre d'etudiant.e.s :', len(noms))
print('Moyenne du groupe :', round(sum(notes) / len(notes), 2))
list(zip(noms, notes))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(noms, notes)
plt.title('Diagramme des notes')
plt.xlabel('Etudiant.e.s')
plt.ylabel('Notes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
