## Flux standards

- STDIN : entrée standard
- STDOUT : sortie standard
- STDERR : sortie d’erreur

Commandes utilisées :
- `>` écrire dans un fichier
- `>>` ajouter à un fichier
- `2>` rediriger les erreurs
- `|` pipe

```python
names = []
grades = []

with open("etudiants.txt", "r", encoding="utf-8") as f:
    for line in f:
        if not line.strip():
            continue
        
        parts = line.split()
        name = " ".join(parts[:-1])   # tout sauf le dernier mot
        grade = float(parts[-1])      # dernier mot = note
        
        names.append(name)
        grades.append(grade)

average = sum(grades) / len(grades)
print(f"Moyenne du groupe: {average:.2f}")
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(names, grades)
plt.axhline(y=average, linestyle="--", label=f"Moyenne: {average:.2f}")
plt.title("Notes des étudiant.e.s avec moyenne")
plt.xlabel("Étudiant.e.s")
plt.ylabel("Notes")
plt.legend()
plt.show()
```
