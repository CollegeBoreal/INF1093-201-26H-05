## Objectif\nLire un fichier texte contenant des étudiants et leurs notes, calculer la moyenne, puis écrire les résultats dans un autre fichier.

```python
with open('etudiants.txt', 'r', encoding='utf-8') as f:\n    lignes = f.readlines()\n\nnotes = []\nfor ligne in lignes:\n    nom, note = ligne.split()\n    notes.append(float(note))\n\nmoyenne = sum(notes) / len(notes)\nprint('Moyenne du groupe :', round(moyenne, 2))
```
