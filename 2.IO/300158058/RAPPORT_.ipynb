```python
noms = []
notes = []

with open("etudiants.txt", "r") as f:
    for ligne in f:
        nom, note = ligne.split()
        noms.append(nom)
        notes.append(int(note))
```

```python
moyenne = sum(notes) / len(notes)
moyenne
round(moyenne, 2)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))  # agrandir le graphique

plt.bar(noms, notes)
plt.axhline(moyenne, linestyle='--', label=f"Moyenne : {round(moyenne, 2)}")

plt.xlabel("Étudiant.e.s")
plt.ylabel("Notes")
plt.title("Notes des étudiant.e.s avec moyenne")

plt.xticks(rotation=45, ha="right")  # rotation des noms
plt.legend()
plt.tight_layout()  # ajuste automatiquement l'espace

plt.show()
```
