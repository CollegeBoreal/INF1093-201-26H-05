# RAPPORT – Traitement des fichiers (Python I/O)

Ce notebook :

• Lit le fichier `etudiants.txt`  
• Calcule la moyenne  
• Génère `resultats.txt`  
• Affiche un diagramme avec la moyenne  


```python

etudiants = []

with open("etudiants.txt", "r", encoding="utf-8") as f:
    for ligne in f:
        ligne = ligne.strip()
        if not ligne:
            continue
        nom, note = ligne.split()
        etudiants.append((nom, int(note)))

etudiants
```

```python

notes = [note for _, note in etudiants]
moyenne = sum(notes) / len(notes)
moyenne
```

```python

seuil = 40

with open("resultats.txt", "w", encoding="utf-8") as f:
    f.write(f"Étudiant.e.s avec note ≥ {seuil} :\n")
    for nom, note in etudiants:
        if note >= seuil:
            f.write(f"{nom} {note}\n")

    f.write("\n")
    f.write(f"Moyenne du groupe : {moyenne:.2f}\n")

print("OK ✅ resultats.txt généré")
```

```python

import matplotlib.pyplot as plt

noms = [nom for nom, _ in etudiants]
notes = [note for _, note in etudiants]

plt.figure(figsize=(9,4))

plt.bar(noms, notes)

# Ligne de la moyenne
plt.axhline(y=moyenne, linestyle='--', label=f"Moyenne : {moyenne:.2f}")

plt.title("Notes des étudiantes avec moyenne")
plt.xlabel("Étudiantes")
plt.ylabel("Notes")

plt.legend()
plt.xticks(rotation=45, ha="right")
plt.tight_layout()

plt.savefig(r"C:\Users\user\Developer\INF1093-201-26H-05\2.IO\300158486\images\diagramme_notes.png", dpi=200)

plt.show()

"diagramme_notes.png"
```
