```python
# Lire les notes
import matplotlib.pyplot as plt
noms = []
notes = []
with open("etudiants.txt", "r", encoding="utf-8") as f:
    for ligne in f:
        nom, note_str = ligne.split()
        noms.append(nom)
        notes.append(float(note_str))
moyenne = sum(notes) / len(notes)
print(f"Moyenne du groupe : {moyenne:.2f}")


plt.figure(figsize=(10,6))
plt.bar(noms, notes, color="skyblue")
plt.ylim(0, max(notes) + 10)
plt.axhline(moyenne, color="green", linestyle="--", linewidth=5, label=f"Moyenne = {moyenne:.1f}",alpha=0.8)
plt.ylabel("Note")
plt.xlabel("Étudiants")
plt.title("Notes des étudiants avec moyenne")
plt.legend(loc="upper left")
plt.tight_layout()
plt.savefig("images/moyenne.png", dpi=150, bbox_inches='tight')
plt.show()
```
