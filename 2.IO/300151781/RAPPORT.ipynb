Ce rapport présente l'analyse visuelle des notes des étudiants traitées par le script `IO.py`.

```python
import matplotlib.pyplot as plt

# Données correspondant à ton fichier etudiants.txt
etudiants = ['lyes', 'john', 'lounes']
notes = [85, 62, 91]

plt.figure(figsize=(8, 5))
colors = ['green' if n >= 60 else 'red' for n in notes]

plt.bar(etudiants, notes, color=colors)

# Mise en forme
plt.axhline(y=60, color='blue', linestyle='--', label='Seuil de réussite (60)')
plt.title('Notes des étudiants')
plt.xlabel('Noms')
plt.ylabel('Note / 100')
plt.ylim(0, 100)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
