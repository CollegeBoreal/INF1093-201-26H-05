```python
import matplotlib.pyplot as plt

noms = ["Maimona", "Bob", "Charlie"]
notes = [99, 62, 91]

# Calcul de la moyenne
moyenne = sum(notes) / len(notes)
print(f"Moyenne du groupe: {moyenne:.2f}")

# Graphique
plt.bar(noms, notes)

# Ligne de moyenne
plt.axhline(moyenne, linestyle='--', label=f"Moyenne: {moyenne:.2f}")

# Axes
plt.xlabel("Étudiant.e.s")
plt.ylabel("Notes")

# Légende
plt.legend()

plt.show()
```
