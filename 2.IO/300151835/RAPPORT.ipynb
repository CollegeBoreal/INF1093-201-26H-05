### Diagramme des notes

**NB :**  
Ce diagramme en barres reprÃ©sente les notes de chaque Ã©tudiant.e.  
Il permet de visualiser rapidement les Ã©carts de performance au sein du groupe.


```python
noms = []
notes = []

with open("etudiants.txt", "r") as f:
    for ligne in f:
        ligne = ligne.strip()
        if not ligne:
            continue

        try:
            nom, note = ligne.split()
            noms.append(nom)
            notes.append(float(note))
        except ValueError:
            print("Ligne invalide :", ligne)

import matplotlib.pyplot as plt

plt.bar(noms, notes)
plt.xlabel("Ã‰tudiant.e.s")
plt.ylabel("Notes")
plt.title("Diagramme des notes")
plt.show()
```
