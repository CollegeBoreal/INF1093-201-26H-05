### Diagramme des notes

**NB :**  
Ce diagramme en barres représente les notes de chaque étudiant.e.  
Il permet de visualiser rapidement les écarts de performance au sein du groupe.


```python
import matplotlib.pyplot as plt

# Lecture des données
noms = []
notes = []

with open("etudiants.txt", "r") as f:
    for ligne in f:
        ligne = ligne.strip()
        if not ligne:
            continue

        try:
            nom, note = ligne.split()
            noms.append(nom)
            notes.append(float(note))
        except ValueError:
            print("Ligne invalide :", ligne)

# Calcul de la moyenne
moyenne = sum(notes) / len(notes) if notes else 0

# Création du diagramme
plt.figure(figsize=(10, 6))
plt.bar(noms, notes, color='skyblue')
plt.axhline(y=moyenne, color='red', linestyle='--', label=f"Moyenne: {moyenne:.2f}")
plt.xlabel("Étudiant.e.s")
plt.ylabel("Notes")
plt.title(f"Notes des étudiant.e.s avec moyenne")
plt.ylim(0, max(notes)+10)
plt.legend()
plt.show()
```
