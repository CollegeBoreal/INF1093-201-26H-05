## 1. Chargement des données

```python
import matplotlib.pyplot as plt
import os
import sys

os.makedirs('images', exist_ok=True)

noms = []
notes = []

try:
    with open('etudiants.txt', 'r') as f:
        for ligne in f:
            ligne = ligne.strip()
            if not ligne:
                continue
            parties = ligne.split()
            if len(parties) == 2:
                try:
                    noms.append(parties[0])
                    notes.append(float(parties[1]))
                except ValueError:
                    print(f"Ligne ignorée : {ligne}", file=sys.stderr)
except FileNotFoundError:
    print("ERREUR: etudiants.txt introuvable.", file=sys.stderr)

print(f"Nombre d'étudiants chargés : {len(noms)}")
for nom, note in zip(noms, notes):
    print(f"  {nom} : {note}")
```

## 2. Calcul de la moyenne

```python
if notes:
    moyenne = sum(notes) / len(notes)
    admis = [(n, note) for n, note in zip(noms, notes) if note >= 60]
    print(f"Moyenne du groupe : {moyenne:.2f}")
    print(f"Étudiants admis (>= 60) : {len(admis)}/{len(noms)}")
else:
    print("Aucune donnée disponible.")
```

## 3. Diagramme des notes

```python
if noms and notes:
    couleurs = ['green' if n >= 60 else 'red' for n in notes]

    fig, ax = plt.subplots(figsize=(10, 5))
    barres = ax.bar(noms, notes, color=couleurs, edgecolor='black')

    ax.axhline(y=moyenne, color='blue', linestyle='--', linewidth=1.5, label=f'Moyenne : {moyenne:.2f}')
    ax.axhline(y=60, color='orange', linestyle=':', linewidth=1.5, label='Seuil de passage : 60')

    for barre, note in zip(barres, notes):
        ax.text(barre.get_x() + barre.get_width() / 2, barre.get_height() + 1,
                f'{note:.0f}', ha='center', va='bottom', fontsize=9)

    ax.set_title('Notes des étudiants', fontsize=14, fontweight='bold')
    ax.set_xlabel('Étudiants')
    ax.set_ylabel('Note (/100)')
    ax.set_ylim(0, 110)
    ax.legend()
    plt.tight_layout()

    plt.savefig('images/diagramme_notes.png', dpi=150)
    print("Figure sauvegardée dans images/diagramme_notes.png")
    plt.show()
```

## 4. Génération du fichier resultats.txt

```python
if noms and notes:
    with open('resultats.txt', 'w') as f:
        f.write('===== RÉSULTATS =====\n\n')
        f.write('Étudiants ayant réussi (note >= 60) :\n')
        for nom, note in zip(noms, notes):
            if note >= 60:
                f.write(f'  {nom} : {note:.1f}\n')
        f.write(f'\nMoyenne du groupe : {moyenne:.2f}\n')
    print("resultats.txt généré avec succès.")
```

## 5. Test des erreurs

```python
try:
    with open('fichier_inexistant.txt', 'r') as f:
        pass
except FileNotFoundError as e:
    print(f"ERREUR capturée : {e}")
    with open('erreurs.txt', 'w') as f_err:
        f_err.write(f"ERREUR: {e}\n")
    print("erreurs.txt généré.")
```
