```python
noms = []
notes = []

with open("etudiants.txt", "r") as f:
    for ligne in f:
        nom, note = ligne.split()
        noms.append(nom)
        notes.append(float(note))

noms, notes
```

```python
moyenne = sum(notes) / len(notes)
moyenne
```

```python
import matplotlib.pyplot as plt

plt.bar(noms, notes)
plt.axhline(y=moyenne, linestyle='--')  # ligne de la moyenne

plt.title("Notes des étudiants")
plt.xlabel("Étudiants")
plt.ylabel("Notes")

plt.text(0, moyenne + 1, f"Moyenne = {moyenne:.1f}")

plt.show()
```
