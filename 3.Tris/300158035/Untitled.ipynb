```python
with open("entree_insertion.txt", "r") as f:
    tab_insertion = list(map(int, f.read().split()))

def tri_insertion(tab):
    for i in range(1, len(tab)):
        cle = tab[i]
        j = i - 1

        while j >= 0 and tab[j] > cle:
            tab[j + 1] = tab[j]
            j -= 1

        tab[j + 1] = cle

tri_insertion(tab_insertion)
print("Insertion :", tab_insertion)
```

```python
with open("entree_shell.txt", "r") as f:
    tab_shell = list(map(int, f.read().split()))

def tri_shell(tab):
    n = len(tab)
    ecart = n // 2

    while ecart > 0:
        for i in range(ecart, n):
            temp = tab[i]
            j = i

            while j >= ecart and tab[j - ecart] > temp:
                tab[j] = tab[j - ecart]
                j -= ecart

            tab[j] = temp

        ecart //= 2

tri_shell(tab_shell)
print("Shell :", tab_shell)
```

```python
with open("entree_quick.txt", "r") as f:
    tab_quick = list(map(int, f.read().split()))

def tri_rapide(tab):
    if len(tab) <= 1:
        return tab

    pivot = tab[len(tab) // 2]

    gauche = []
    milieu = []
    droite = []

    for x in tab:
        if x < pivot:
            gauche.append(x)
        elif x == pivot:
            milieu.append(x)
        else:
            droite.append(x)

    return tri_rapide(gauche) + milieu + tri_rapide(droite)

tab_trie = tri_rapide(tab_quick)
print("Quick :", tab_trie)
```
