Ce notebook présente trois algorithmes de tri faits en Python avec lecture depuis des fichiers texte.

## Algorithmes utilisés
- Tri par insertion
- Tri de Shell
- Tri rapide (Quick Sort)

```python
tableau = [22, 5, 17, 9, 1, 14, 3, 11]
def tri_insertion(tab):
    for i in range(1, len(tab)):
        valeur = tab[i]
        j = i - 1
        while j >= 0 and tab[j] > valeur:
            tab[j + 1] = tab[j]
            j -= 1
        tab[j + 1] = valeur
    return tab
print(tri_insertion(tableau))
```
