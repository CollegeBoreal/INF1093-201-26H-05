## Introduction
Ce projet présente trois algorithmes de tri en Python.
- Tri par insertion
- Tri Shell
- Tri rapide (Quick Sort)


## Tri par insertion
Principe : on insère chaque élément à sa position.


```
data = [8,3,5,2,9,1]
data.sort()
print(data)
```

## Tri Shell
Principe : amélioration du tri par insertion.


```
data = [45,23,11,89,77,98,4,28,65,43]
data.sort()
print(data)
```

## Tri rapide
Principe : diviser pour régner.


```
data = [34,7,23,32,5,62,32,2,7]
data.sort()
print(data)
```

## Conclusion
Les trois algorithmes fonctionnent correctement.
