## Objectif
- Lire des donnÃ©es depuis un fichier texte
- Appliquer un algorithme de tri
- Afficher le rÃ©sultat

Algorithmes Ã©tudiÃ©s :
1. Tri par insertion
2. Tri de Shell

```python
# Lecture du fichier
with open("insertion/entree_insertion.txt", "r") as f:
    tab = list(map(int, f.read().split()))

def tri_insertion(tab):
    for i in range(1, len(tab)):
        cle = tab[i]
        j = i - 1

        while j >= 0 and tab[j] > cle:
            tab[j + 1] = tab[j]
            j -= 1

        tab[j + 1] = cle

tri_insertion(tab)
print("RÃ©sultat Insertion :", tab)
```

```python
# Lecture du fichier
with open("shell/entree_shell.txt", "r") as f:
    tab = list(map(int, f.read().split()))

def tri_shell(tab):
    n = len(tab)
    ecart = n // 2

    while ecart > 0:
        for i in range(ecart, n):
            temp = tab[i]
            j = i

            while j >= ecart and tab[j - ecart] > temp:
                tab[j] = tab[j - ecart]
                j -= ecart

            tab[j] = temp

        ecart //= 2

tri_shell(tab)
print("RÃ©sultat Shell :", tab)
```

```python
# Lecture du fichier
with open("quick/entree_quick.txt", "r") as f:
    tab = list(map(int, f.read().split()))

def tri_rapide(tab):
    if len(tab) <= 1:
        return tab

    pivot = tab[len(tab) // 2]

    gauche = []
    milieu = []
    droite = []

    for x in tab:
        if x < pivot:
            gauche.append(x)
        elif x == pivot:
            milieu.append(x)
        else:
            droite.append(x)

    return tri_rapide(gauche) + milieu + tri_rapide(droite)

tab_trie = tri_rapide(tab)
print("RÃ©sultat Quick :", tab_trie)
```
